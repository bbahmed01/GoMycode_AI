```python
def loadDataset(filename, split, trainingSet=[] , testSet=[]):

    with open(filename, 'r') as csvfile:

        lines = csv.reader(csvfile)
        dataset = list(lines)
        print('dataset length : ',len(dataset))
        for x in range(len(dataset)):
    
            l = []
            for y in range(5):
                
                if dataset[x][y]>'0.0':
                    l.append(dataset[x][y])
                    
            if random.random() < split:
                    
                trainingSet.append(l)

            else:
                testSet.append(l)
                        
trainingSet=[]

testSet=[]

loadDataset('iris.data.txt', 0.66, trainingSet, testSet)

print('trainingSet == testSet : ', trainingSet == testSet)

print ('Train: ' + repr(len(trainingSet)))

print ('Test: ' + repr(len(testSet)) )
```

```python
def euclideanDistance(instance1, instance2, length):
    distance = 0
    for i in range(len(instance1[0])):
        for k in range(length):
            #print(instance1[i][k],' | ',instance2[k],' ', i , ' ', k)
            d1 = float(instance1[i][k])
            d2 = float(instance2[k])
        distance = distance +  math.sqrt(math.pow(d1-d2,2))
    return distance


        
data1 = trainingSet
data2 = testSet[0]

distance = euclideanDistance(data1, data2, 4)


print ('Distance: ' + repr(distance))
```

```python
def getNeighbors(trainingSet, testInstance, k):

    distances = []
    neighbors_of_x = []
    
    length = len(testInstance)-1
    for x in range(len(trainingSet)-1):

        dist = euclideanDistance(testInstance, trainingSet[x], k)
        
        distances.append((trainingSet[x], dist))

        distances.sort(key=operator.itemgetter(1))

        
    for d in range(len(distances)):
        neighbors = []
        for i in range(k):
            neighbors.append(distances[d][0])


        neighbors_of_x.append(neighbors)
        
    return neighbors, neighbors_of_x


trainSet = trainingSet
#print('trainSet = trainingSet : ',trainSet == trainingSet)

testInstance = testSet
#print(' testInstance = testSet : ', testInstance == testSet)

neighbors, neighbors_of_x = getNeighbors(trainSet, testInstance, 4)


print('neighbors : ', neighbors)

print('neighbors of x : ', neighbors_of_x)
```

```python
def getResponse(neighbors):

    classVotes = {}

    for x in range(len(neighbors)):

        response = neighbors[x][4] #complete with appropriate number
        if response in classVotes.keys():
            classVotes[response] = classVotes[response] + 1
        else:
            classVotes[response] = 1

    if response in classVotes:
            sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
            
    return sortedVotes[0][0]
    
"""neighbors = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]"""

response = getResponse(neighbors)

print(response)
```

```python
def getAccuracy(testSet, predictions):
    counter = -1
    correct = 0
    for items in testSet:
        counter = counter + 1
        #print(items[4],' ',predictions[counter])
        if items[4] == predictions[counter]:
            correct = correct + 1
    return (correct/float(len(testSet))) * 100.0


testSet1 = [[1,1,1,1,'a'], [2,2,2,2,'a'], [3,3,3,3,'b']]

predictions1 = ['a', 'a', 'a']

accuracy = getAccuracy(testSet1, predictions1)

print(accuracy,"%")
```

```python
def ManhattanDistance(instance1, instance2, length):
    distance = 0
    for i in range(len(instance1[0])):
        for k in range(length):
            #print(instance1[i][k],' | ',instance2[k],' ', i , ' ', k)
            d1 = float(instance1[i][k])
            d2 = float(instance2[k])
        distance = distance +  abs(d1-d2)
    return distance

def getNeighborsM(trainingSet, testInstance, k):

    distances = []
    neighbors_of_x = []
    
    length = len(testInstance)-1
    for x in range(len(trainingSet)-1):

        dist = ManhattanDistance(testInstance, trainingSet[x], k)
        
        distances.append((trainingSet[x], dist))

        distances.sort(key=operator.itemgetter(1))

        
    for d in range(len(distances)):
        neighbors = []
        for i in range(k):
            neighbors.append(distances[d][0])


        neighbors_of_x.append(neighbors)
        
    return neighbors, neighbors_of_x

```
